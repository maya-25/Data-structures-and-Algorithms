Binary search Tree Implementation

```python
class BinaryserachTree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    def child(self,data):
        if data==self.data:
            return
        if data<self.data:
            #add data in left subtree
            if self.left:
                self.left.child(data)
            else:
                self.tree=BinaryserachTree(data)
        else:
            #add data in right subtree
            if self.left:
                self.left.child(data)
            else:
                self.tree=BinaryserachTree(data)
    
    
    def in_order_traversal(self):
        elements=[]
        #visit the left tree first
        if self.left:
            elements+=self.left.in_order_traversal()
        #visit the root node
        elements.append(self.data)
        
        #visit the right tree
        if self.right:
            elements+=self.right.in_order_traversal()
        return elements
    def search(self,val):
        if self.data==val:
            return True
        if val<self.data:
            #val might be in left sub tree
            if self.left:
                self.left.search(val)
            else:
                return False
        if val>self.data:
            #val might be in right sub tree
            if self.right:
                self.right.search(val)
            else:
                return False
    def find_min(self):
        if self.left is None:
            return self.data
        return self.left.find_min()
    def find_max(self):
        if self.right is None:
            return self.data
        return self.right.find_max()
    def calculate_sum(self):
        left_sum=self.left.calculate_sum() if self.left else 0
        right_sum=self.right.calculate_sum() if self.right else 0
        return self.data + left_sum + right_sum
    
    def post_order_traversal(self):
        elements=[]
        #visit the left tree first
        if self.left:
            elements+=self.left.post_order_traversal()
        
        
        #visit the right tree
        if self.right:
            elements+=self.right.post_order_traversal()
        return elements
        
        #visit the root node
        elements.append(self.data)
        
    def pre_order_traversal(self):
        #visit the root node
        elements=[self.data]
    
        #visit the left tree first
        if self.left:
            elements+=self.left.pre_order_traversal()
        
        
        #visit the right tree
        if self.right:
            elements+=self.right.pre_order_traversal()
        return elements
    
def build_tree(elements):
    root=BinaryserachTree(elements[0])
    
    
    for i in range (1,len(elements)):
        root.child(elements[i])
    return root
    
        
    
if __name__=='__main__':
    numbers=[12,13,15,11,10,9,200,190]
    number_tree=build_tree(numbers)
    print(number_tree.search(15))
#     numbers_tree = build_tree(numbers)
    print("Input numbers:",numbers)
    print("Min:",number_tree.find_min())
    print("Max:",number_tree.find_max())
    print("Sum:", number_tree.calculate_sum())
    print("In order traversal:", number_tree.in_order_traversal())
    print("Pre order traversal:", number_tree.pre_order_traversal())
    print("Post order traversal:", number_tree.post_order_traversal())
```
